```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_dataset
import warnings
warnings.filterwarnings('ignore')
plt.style.use('dark_background')
sns.set_theme(style="darkgrid", rc={"axes.facecolor": "#1e1e1e", "figure.facecolor": "#121212", "text.color": "white", "axes.labelcolor": "white", "xtick.color": "white", "ytick.color": "white"})
```

## 1. Load Data Subset

```python
print("Downloading a small subset for EDA...")
dataset = load_dataset("Rajarshi-Roy-research/Defactify_Image_Dataset", split="train[:5000]")
print(f"Loaded {len(dataset)} images.")
```

## 2. Graph 1: The Massive Class Imbalance (Head A)


```python

labels_a = ["Authentic (Real)" if x == 0 else "AI-Generated" for x in dataset['label']]

plt.figure(figsize=(8, 5), dpi=150)
ax = sns.countplot(x=labels_a, palette=["#4CAF50", "#F44336"])
plt.title("Class Imbalance: Authentic vs AI-Generated (Training Data)", fontsize=14, pad=15, weight='bold')
plt.ylabel("Number of Images", fontsize=12)
plt.xlabel("")

total = len(labels_a)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height() + 50
    ax.annotate(percentage, (x, y), size=12, weight='bold')

plt.tight_layout()
plt.savefig("eda_imbalance.png", transparent=True)
plt.show()
```

## 3. Graph 2: Generator Distribution (Head B)

```python
generator_map = {0: "Real", 1: "SD 2.1", 2: "SDXL", 3: "SD 3", 4: "DALL·E 3", 5: "Midjourney"}
labels_b = [generator_map[x] for x in dataset['multi_label']]

plt.figure(figsize=(10, 5), dpi=150)
ax = sns.countplot(y=labels_b, order=list(generator_map.values()), palette="viridis")
plt.title("Distribution of Image Generators in Dataset", fontsize=14, pad=15, weight='bold')
plt.xlabel("Number of Images", fontsize=12)
plt.ylabel("")

plt.tight_layout()
plt.savefig("eda_generators.png", transparent=True)
plt.show()
```

## 4. Visual Comparison Grid

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8), dpi=150)
fig.suptitle("Can you tell what is real? (Visual Indistinguishability)", fontsize=16, weight='bold', y=1.02)

samples = []
for i in range(6):
    for item in dataset:
        if item['multi_label'] == i:
            samples.append((item['image'], generator_map[i]))
            break

for ax, (img, label) in zip(axes.flatten(), samples):
    ax.imshow(img)
    ax.set_title(f"Source: {label}", fontsize=12, color='white', pad=10, backgroundcolor='#333333')
    ax.axis('off')

plt.tight_layout()
plt.savefig("eda_grid.png", transparent=True)
plt.show()
```
